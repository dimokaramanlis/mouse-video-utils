# Apply Keypoint MoSeq to new data

We will first import the libraries needed. If you get an error, check that you're running the notebook inside the environment where DeepLabCut is installed.

```python
import keypoint_moseq as kpms
import os, glob, h5py, kptmoseq_fun    
from jax_moseq.utils import set_mixed_map_iters
set_mixed_map_iters(4)
```

Below we specify the data folder for the project. We need a DLC folder as well.

```python
project_dir = 'S:\\ElboustaniLab\\#SHARE\\Analysis\\KeypointMoSeq\\MoSeq_Training'
model_name  = '2024_07_05-14_14_10'

# load the most recent model checkpoint and pca object
model  = kpms.load_checkpoint(project_dir, model_name)[0]
config = lambda: kpms.load_config(project_dir)
# load new data (e.g. from deeplabcut)
new_data_path = os.path.join(project_dir, 'data_run') # can be a file, a directory, or a list of files
```

```python
# make the list of results
results_files  = list(glob.glob(os.path.join(project_dir, model_name,'results','*.csv')))
```

```python
for file in glob.glob(os.path.join(new_data_path, '*.csv')):
    
    directory, filename = os.path.split(file)
    
    if kptmoseq_fun.contains(filename[:25], results_files):
        print(filename[:25] + ' session already analyzed')
        continue
        
    print('Current session: ' + filename[:25])
    coordinates, confidences, bodyparts = kpms.load_keypoints(file, 'deeplabcut')
    data, metadata = kpms.format_data(coordinates, confidences, **config())

    # apply saved model to new data
    results = kpms.apply_model(model, data, metadata, project_dir, model_name, **config())
    
    # optionally rerun `save_results_as_csv` to export the new results
    kpms.save_results_as_csv(results, project_dir, model_name)
```
